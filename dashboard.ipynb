```python
import pandas as pd

def purge_dataset(input_file='all_matches.csv', output_file='all_matches_clean.csv'):
    print(f"üßπ Starting Data Purge on {input_file}...")
    
    # 1. Load Data
    df = pd.read_csv(input_file, low_memory=False)
    
    # Clean column names (strip spaces, lowercase)
    df.columns = df.columns.str.strip().str.lower()
    
    # 2. Define Filters
    # CUTOFF DATE: Jan 1, 2016
    CUTOFF_DATE = '2016-01-01'
    
    # TOP 10 TEAMS (The Elite List)
    TOP_10 = [
        'India', 'Australia', 'Pakistan', 'South Africa', 'New Zealand',
        'Sri Lanka', 'England', 'Bangladesh', 'Afghanistan', 'West Indies'
    ]
    
    initial_rows = len(df)
    
    # 3. Apply Date Filter
    df['start_date'] = pd.to_datetime(df['start_date'], errors='coerce')
    df = df[df['start_date'] >= CUTOFF_DATE]
    rows_after_date = len(df)
    
    print(f"   üìÖ Date Filter (Post-2016): Removed {initial_rows - rows_after_date} old rows.")
    
    # 4. Apply Team Filter (Strict: Both Teams must be Top 10)
    # Identify correct column names for teams
    if 'batting_team' in df.columns:
        t1, t2 = 'batting_team', 'bowling_team'
    else:
        t1, t2 = 'team_bat_1', 'team_bat_2'
        
    df = df[
        (df[t1].isin(TOP_10)) & 
        (df[t2].isin(TOP_10))
    ]
    
    final_rows = len(df)
    print(f"   üõ°Ô∏è Team Filter (Top 10 Only): Removed {rows_after_date - final_rows} associate/weak rows.")
    
    # 5. Save the Fresh File
    df.to_csv(output_file, index=False)
    
    print("-" * 40)
    print(f"‚úÖ SUCCESS! Cleaned dataset saved as '{output_file}'")
    print(f"üìä Final Dataset Size: {final_rows} rows (Reduced by {initial_rows - final_rows})")
    print("-" * 40)

# Run the function
purge_dataset(input_file='data/FINAL_ODI_MASTER.csv')
```
